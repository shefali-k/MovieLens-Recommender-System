```python
#1. Draw scatter plots for casual vs registered users
plt.scatter(df["casual"], df["registered"], c='y', alpha=0.5)
plt.xlabel("Casual Users")
plt.ylabel("Registered Users");
plt.show()
```

```python
#2. Draw a histogram showing bike sharing data for various seasons
n, bins, patches = plt.hist(x=df["season"], bins='auto', color='#0504aa',
                            alpha=0.7, rwidth=0.85)
plt.grid(axis='y', alpha=0.75)
plt.xlabel('Season')
plt.ylabel('Frquency')
```

```python
#3. Draw a histogram for various weather situations.
n, bins, patches = plt.hist(x=df["weathersit"], bins='auto',
                            alpha=0.7, rwidth=0.85)
plt.grid(axis='y', alpha=0.75)
plt.xlabel('Weather Situation')
plt.ylabel('Frquency')
```
